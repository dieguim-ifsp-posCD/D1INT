```python
from sklearn.cluster import KMeans
from sklearn import preprocessing

import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline 

import warnings
warnings.filterwarnings('ignore')
```

```python
data = pd.read_csv('Mall_Customers.csv')
```

```python
data.rename(index=str, columns={'Annual Income (k$)': 'Income',
                              'Spending Score (1-100)': 'Score'}, inplace=True)
data.head(10)
```

```python
data.shape
```

```python
plt.scatter(data['Income'], data['Score'])
plt.xlabel("Annual Income (k$)")
plt.ylabel("Spending Score (1-100)")
```

```python
X = data.copy()
X.drop(['CustomerID'], axis=1, inplace=True)
X.loc[X['Gender'] == 'Male', 'Gender'] = 0
X.loc[X['Gender'] == 'Female', 'Gender'] = 1
X.head()
```

```python
kmeans = KMeans(3, random_state=0)
kmeans.fit(X)

plt.figure(figsize=(10,6))
plt.scatter(data['Income'], data['Score'], c=kmeans.fit_predict(X), cmap='rainbow')
plt.xlabel("Annual Income (k$)")
plt.ylabel("Spending Score (1-100)")
plt.title('KMeans with 3 Clusters')
plt.show()
```

```python
kmeans = KMeans(5, random_state=0)
kmeans.fit(X)

plt.figure(figsize=(10,6))
plt.scatter(data['Income'], data['Score'], c=kmeans.fit_predict(X), cmap='rainbow')
plt.xlabel("Annual Income (k$)")
plt.ylabel("Spending Score (1-100)")
plt.title('KMeans with 5 Clusters')
plt.show()
```

```python
kmeans = KMeans(3, random_state=0)
kmeans.fit(X)

plt.figure(figsize=(10,6))
plt.scatter(data['Age'], data['Score'], c=kmeans.fit_predict(X), cmap='rainbow')
plt.xlabel("Age")
plt.ylabel("Spending Score (1-100)")
plt.title('KMeans with 3 Clusters')
plt.show()
```

```python
kmeans = KMeans(5, random_state=0)
kmeans.fit(X)

plt.figure(figsize=(10,6))
plt.scatter(data['Age'], data['Score'], c=kmeans.fit_predict(X), cmap='rainbow')
plt.xlabel("Age")
plt.ylabel("Spending Score (1-100)")
plt.title('KMeans with 3 Clusters')
plt.show()
```
